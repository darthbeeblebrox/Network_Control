```python
import TestNetworks as tn
import DigraphConvert as dc
import matplotlib.pyplot as plt
import Maximal_matching as mm
import networkx as nx

%matplotlib inline

g=nx.Graph()
edges=[
        [(1,0), (0,0)],
        [(1,1), (0,0)],
        [(1,2), (0,0)],
        ]

for ii in edges:
    g.add_node(ii[0],bipartite=0)
    g.add_node(ii[1],bipartite=1)

g.add_edges_from(edges)
mm.plotGraph(g)

#print(g.nodes(data=True))

all_matches=mm.enumMaximumMatching(g)
print("The matches for test network",1,"are",all_matches)
```

```python
from timeit import default_timer as timer
import operator as op

starts = []
ends = []
maxSize = 23

for i in range(1,maxSize):
    G = nx.random_graphs.erdos_renyi_graph(i,0.2,directed=True)
    starts.append(timer())
    tempcycles = list(nx.simple_cycles(G))
    ends.append(timer())

plt.plot(range(1,maxSize),list(map(op.sub, ends, starts)))
plt.ylabel('Time of cycle algorithm (log(s))')
plt.yscale('log')
plt.xlabel('Size of random graph')
plt.show
```
