```python
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np
import tarjan as tj

# network1 = {1:[2],2:[1,5],3:[4],4:[3,5],5:[6],6:[7],7:[8],8:[6,9],9:[]}
# networkWang = {1:[16],2:[3,4],3:[],4:[],5:[6],6:[7],7:[8],8:[5],9:[10,12],10:[11],11:[10],12:[],13:[9,14],14:[15],15:[13],16:[5,9,13]}

# tj.tarjan(networkWang)

G = nx.read_graphml('elegansGraphML.txt',node_type=int)

CELdict = {}
for edge in G.edges():
    if edge[0] in CELdict:
        CELdict[edge[0]].append(edge[1])
    else:
        CELdict[edge[0]] = [edge[1]]

tresult = tj.tarjan(CELdict)
print(tresult)
print(len(tresult[17]))
print('******')
newDict = {}
for n in tresult[17]:
    oldNodes = CELdict[n]
    for on in oldNodes:
        if on in tresult[17]:
            if n in newDict:
                newDict[n].append(on)
            else:
                newDict[n] = [on]
print(newDict)
tresult2 = tj.tarjan(newDict)
print(tresult2)
for q in newDict:
    print(len(newDict[q]))
```
