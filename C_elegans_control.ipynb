# Build the C elegans network
Uses a graph_ml text file downloaded from https://elegans.herokuapp.com/#.

```python
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
```

```python
#Import the network and visualize it
# Note that this is a directed graph!

G = nx.read_graphml('elegansGraphML.txt')

nx.draw(G,node_size=20,width=0.1)
```

# Simple network statistics
Using built-in functions from networkx

```python
#This command takes a really long time!
# print('The average node connectivity is ',nx.average_node_connectivity(G))

#This fails for a directed graph
#print('The graph is connected:',nx.is_connected(G))

print('The graph is strongly connected:',nx.is_strongly_connected(G))
print('The graph is weakly connected:',nx.is_weakly_connected(G))

dList = nx.degree_histogram(G)
#dHist = np.histogram(dList)
plt.hist(dList)
plt.title("Histogram of degree distributions")
```

```python
Gcolor = nx.coloring.greedy_color(G)

#nx.draw(G,node_color=Gcolor)

D = nx.degree_centrality(G)
plt.bar(range(len(D)), D.values(), align='center')
plt.title("Degree centrality for each no")
#plt.xticks(range(len(D)), D.keys())

plt.show()
```
